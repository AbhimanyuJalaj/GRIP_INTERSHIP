```python
# importing library

import pandas as pd
```

```python
# loading dataset

import requests

from io import StringIO

fileId = "https://drive.google.com/file/d/11Iq7YvbWZbt8VXjfm06brx66b10YiwK-/view?usp=sharing".split('/')[-2]

downurl = 'http://drive.google.com/uc?export=download&id=' + fileId

csvFile = StringIO(requests.get(downurl).text)

df = pd.read_csv(csvFile)

df
```

```python
# assigning numerical value for species.

numValueToFeatures = dict([('Iris-setosa',0),('Iris-versicolor',1),('Iris-virginica',2)])

df.Species = [numValueToFeatures[i] for i in df.Species]

df
```

```python
# Apllying decision tree algorithm

from sklearn.tree import DecisionTreeClassifier

Tree =DecisionTreeClassifier()

Tree.fit(df[df.columns[1:5]],df['Species'])
```

```python
# creating tree for visualization.

from sklearn.externals.six import StringIO  

from IPython.display import Image  

from sklearn.tree import export_graphviz

import pydotplus


dot_data = StringIO()

export_graphviz(Tree, out_file=dot_data, feature_names=df[df.columns[1:5]].columns.tolist(),  
                filled=True, rounded=True,
                special_characters=True)

with open('dot_data.xml','w') as f :
    f.write(dot_data.getvalue())
    
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 

Image(graph.create_png())
```
