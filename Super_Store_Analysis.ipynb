```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sn
```

```python
# loading dataset

import requests

from io import StringIO

fileId = "https://drive.google.com/file/d/1lV7is1B566UQPYzzY8R2ZmOritTW299S/view".split('/')[-2]

downurl = 'http://drive.google.com/uc?export=download&id=' + fileId

csvFile = StringIO(requests.get(downurl).text)

df = pd.read_csv(csvFile)

df
```

```python
df['Country'].nunique()
```

```python
del df['Country']

df
```

```python
df['Cost'] = df['Sales'] - df['Profit']

df['PercentProfit'] = (df['Profit']/df['Cost'])*100

df
```

```python
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()

plt.title('Total Profit and Sales')

plt.show()
```

```python
plt.figure(figsize=(12,10))

df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")

plt.show()
```

```python
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()

plt.title('Total Profit and Sales')

plt.show()
```

```python
df.groupby('Region')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sn.countplot(x="Sub-Category",hue="Region",data=df)
plt.show()
```
