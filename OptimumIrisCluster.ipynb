```python
# importing important library

import pandas as pd

import matplotlib.pyplot as plt
```

```python
# loading dataset
import requests

from io import StringIO

fileId = "https://drive.google.com/file/d/11Iq7YvbWZbt8VXjfm06brx66b10YiwK-/view?usp=sharing".split('/')[-2]

downurl = 'http://drive.google.com/uc?export=download&id=' + fileId

csvFile = StringIO(requests.get(downurl).text)

df = pd.read_csv(csvFile)

df
```

# As we know that clustering is unsupervised algorithm, we need only features.
So, our next step to seprate features.

```python
df_features = df.iloc[:, 1:5 ]

df_features
```

# We used elbow method to determine optimal number of cluster.

1) We calculated the sum of square distance of sample to their closest cluster for given value of K.

2) there is optimum number of cluster appears when the ratio of sum of square distance of sample at value of K-1 is to that of K-2 is less than 0.55.

```python
from sklearn.cluster import KMeans

squareDistanceSum = []

for k in range(1,10) :
    
    km = KMeans(n_clusters = k)
    
    km = km.fit(df_features)
    
    squareDistanceSum.append(km.inertia_)
    
    if k > 1 : 
        ratio = squareDistanceSum[k-1]/squareDistanceSum[k-2]
        if ratio < 0.55 :
            optimalvalue = k
            
print("Optimum number of cluster is : ",optimalvalue)
```

```python
# plotting graph for value of squareDistanceSum and K to show elbow.

plt.plot(range(1,10),squareDistanceSum)

plt.xlabel("K")
plt.ylabel("squareDistanceSum")

plt.title("Determing K by elbow of graph")

plt.show()
```

# Applying K means clustering algo to predict the output.

you see that ouput by algorithm is 0,1,2 .

so, you face difficulty in mapping that whether which number represent which species of flower.

```python
kmeans = KMeans(n_clusters = optimalvalue)

y_kmeans = kmeans.fit_predict(df_features)

print(len(y_kmeans),type(y_kmeans))

df_label = pd.DataFrame()

df_label['Species'] = df['Species']

df_label['numValue'] = pd.Series(y_kmeans.tolist())

df_label
```

```python
# Grouping all prediction done by clustering algorithm.

c = df_label.groupby(['Species','numValue'])

distinctPredCount = c.size().tolist()

c = pd.DataFrame(c)

c['count'] = pd.Series(distinctPredCount)

c
```

```python
#deleting unused coloumn

del c[1]

c = c.rename(columns={0:'Species,Predicted_Value'})

c
```

```python
# finding which value is associated with which species by the fact that the value associated with maximum number 
# of sample predicted by the algorithm is the value associated to it.

setosaSample = 0
versicolorSample = 0
virginicaSample = 0
setosaValue = 0
versicolorValue = 0
virginicaValue = 0
for i in range(len(c.index)) :
    if c['Species,Predicted_Value'][i][0] == 'Iris-setosa' :
        if(setosaSample<c['count'][i]) :
            setosaSample = c['count'][i]
            setosaValue = c['Species,Predicted_Value'][i][1]
    elif c['Species,Predicted_Value'][i][0] == 'Iris-versicolor' :
        if(versicolorSample<c['count'][i]) :
            versicolorSample = c['count'][i]
            versicolorValue = c['Species,Predicted_Value'][i][1]
    else :
        if(virginicaSample<c['count'][i]) :
            virginicaSample = c['count'][i]
            virginicaValue = c['Species,Predicted_Value'][i][1]
        
n = "\n"        
print("setosaValue = ",setosaValue,n,"versicolorValue = ",versicolorValue,n,"virginicaValue = ",virginicaValue)
```

```python
# the cluster centers

kmeans.cluster_centers_
```

```python
# plot on sepal features.

x = df_features.values

plt.scatter(x[y_kmeans == setosaValue, 0], x[y_kmeans == setosaValue, 1], s = 50, c = 'red', label = 'Iris-setosa')

plt.scatter(x[y_kmeans == versicolorValue, 0], x[y_kmeans == versicolorValue, 1], s = 50, c = 'blue', label = 'Iris-versicolor')

plt.scatter(x[y_kmeans == virginicaValue, 0], x[y_kmeans == virginicaValue, 1], s = 50, c = 'green', label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")

plt.title("Sepal length vs width")

plt.legend()
```

# It is impossible to visualize all four feature point in 2D graph. That's why we used principle component analysis to convert 4 features into two features.

```python
from sklearn.decomposition import PCA

model = PCA(n_components = 2)

model.fit(df_features)

transformedFeatures = model.transform(df_features)

transformedFeatures
```

```python
# transforming label by PCA

transformedLabels = model.transform(kmeans.cluster_centers_)
transformedLabels
```

```python
# graph for principle component.

x = transformedFeatures

plt.scatter(x[y_kmeans == setosaValue, 0], x[y_kmeans == setosaValue, 1], s = 50, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == versicolorValue, 0], x[y_kmeans == versicolorValue, 1], s = 50, c = 'blue', label = 'Iris-versicolor')
plt.scatter(x[y_kmeans == virginicaValue, 0], x[y_kmeans == virginicaValue, 1], s = 50, c = 'green', label = 'Iris-virginica')


plt.scatter(transformedLabels[:, 0], transformedLabels[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.xlabel("Principle Component 1")
plt.ylabel("Principle Component 2")

plt.title("plot of principle components")

plt.legend()
```
